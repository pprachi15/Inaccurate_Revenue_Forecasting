# Converting dtypes from object to category

```python
cat_cols = ['Month_Name','Product', 'Model_Version']
for n in cat_cols:
    df[n] = df[n].astype('category')
```

# Checking numeric columns 

```python
num_cols = ['Forecast_Revenue', 'Actual_Revenue',	'Absolute_Error',	'Absolute_Error_Pct',	'Marketing_Spend',	'Macro_Demand_Index']
for x in num_cols:
    df[x] = pd.to_numeric(df[x], errors= 'coerce')
```

```python
df.info()
```

# Create Engineered Features

```python
print("""
APE : Absolute Percent Error, measuring how large the forecasting mistake is relative to actual revenue for a single period.

MAPE : Mean Absolute Percent Error, showing the average size of forecasting mistakes across all periods.

MKT Ratio : Marketing spend as a share of revenue, indicating how much marketing is invested for every dollar earned.

Bias : The directional forecasting error that shows whether predictions tend to be consistently higher or lower than actuals.
""")
```

```python
df['Bias'] = df['Forecast_Revenue'] - df['Actual_Revenue']
df['APE'] = (df['Bias'].abs() / df['Actual_Revenue']).clip(0, 1)
df['Mkt_Ratio'] = df['Marketing_Spend'] / df['Actual_Revenue']
```

```python
df.info()
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Month"].astype(str) + "-01")
```

# Defining Colors

```python
# JPMorgan-style color palette
JPM_BROWN = "#5A4A42"
JPM_NAVY = "#0D1F2D"
JPM_STEEL = "#4A6572"
JPM_TEAL = "#2C7A7B"
JPM_GRAY = "#A8A8A8"
JPM_CHARCOAL = "#1A1A1A"
JPM_CREAM = "#F7F4EF"

plt.style.use("default")

plt.rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "axes.edgecolor": JPM_GRAY,
    "axes.labelcolor": JPM_CHARCOAL,
    "xtick.color": JPM_CHARCOAL,
    "ytick.color": JPM_CHARCOAL,
    "grid.color": JPM_GRAY,
    "font.size": 11,
    "axes.titlesize": 13,
    "axes.titleweight": "bold"
})
```

# Portfolio Forecast Accuracy Over Time

```python
portfolio = df.groupby("Date", as_index = False).agg(
    Actual_Revenue = ("Actual_Revenue", "sum"),
    Forecast_Revenue = ("Forecast_Revenue", "sum")
)
```

```python
# Convert revenue to millions
portfolio["Actual_M"] = portfolio["Actual_Revenue"] / 1_000_000
portfolio["Forecast_M"] = portfolio["Forecast_Revenue"] / 1_000_000

plt.figure(figsize=(10,5))

plt.plot(
    portfolio["Date"],
    portfolio["Actual_M"],
    label="Actual",
    color=JPM_NAVY,
    linewidth=2.5,
)

plt.plot(
    portfolio["Date"],
    portfolio["Forecast_M"],
    label="Forecast",
    color=JPM_BROWN,
    linewidth=2.5,
    linestyle="--"
)

plt.title("Actual vs Forecast Revenue (Millions USD)", color=JPM_CHARCOAL)
plt.xlabel("Month", color=JPM_CHARCOAL)
plt.ylabel("Revenue (Millions USD)", color=JPM_CHARCOAL)

plt.grid(alpha=0.3)
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```

# Which product is hurting the forecasting accuracy the most?

```python
# Compute MAPE (Mean Absolute Percentage Error) by Product
by_product = (
    df.groupby("Product")
     .agg(
         MAPE = ("APE", "mean"),
         Bias_pct = ("Bias", lambda s: (s / df.loc[s.index, "Actual_Revenue"]).mean())
     )
     .reset_index()
)
by_product.round(3)
```

```python
plt.figure(figsize=(7,5))
plt.bar(
    by_product["Product"],
    by_product["MAPE"],
    color = [JPM_NAVY, JPM_BROWN, JPM_TEAL]
)
plt.title("Forecast Accuracy by Product", color = JPM_CHARCOAL)
plt.ylabel("Mean Absolute percentage Error", color = JPM_CHARCOAL)
plt.grid(axis = 'y', alpha=0.25)
plt.tight_layout()
plt.show()
```

# Can we improve the forecast with a simple calibration layer?

```python
calib_factors = (
    df.groupby(["Product", "Month"])
    .agg(mean_err_pct=("Absolute_Error_Pct", "mean"))
    .reset_index()
)
calib_factors.head()
```

```python
# merging calibration factors into your dataset
df_cal = df.merge(calib_factors, on=["Product", "Month"], how= "left")
```

```python
# Apply calibrated forecast
df_cal["Forecast_Calibrated"] = df_cal["Forecast_Revenue"] * (1 - df_cal["mean_err_pct"])
```

```python
# Recompute calibrated Bias and APE
df_cal["Bias_Calibrated"] = df_cal["Forecast_Calibrated"] - df_cal["Actual_Revenue"]

df_cal["APE_Calibrated"] = (
    df_cal["Bias_Calibrated"].abs() / df_cal["Actual_Revenue"]
).clip(0,1)
```

```python
# Compare overall performance
before = df["APE"].mean()
after = df_cal["APE_Calibrated"].mean()
print("MAPE before calibration: ", round(before, 3))
print("MAPE after calibration: " , round(after, 3))
print("Improvement (pp): ", round(before - after, 3)) 
```

```python
# compare by product
compare_products = (
    df_cal.groupby("Product")
    .agg(
        MAPE_before = ("APE", "mean"),
        MAPE_after = ("APE_Calibrated", "mean")
    )
    .round(3)
)
compare_products["Improvement_pp"] = compare_products["MAPE_before"] - compare_products["MAPE_after"]
compare_products
```

```python
plt.figure(figsize=(7,5))

plt.bar(compare_products.index, compare_products["MAPE_before"],
        color=JPM_BROWN, alpha=0.7, label="Before")

plt.bar(compare_products.index, compare_products["MAPE_after"],
        color=JPM_TEAL, alpha=0.7, label="After")

plt.title("MAPE Before and After Calibration", color=JPM_CHARCOAL)
plt.ylabel("MAPE", color=JPM_CHARCOAL)
plt.grid(axis="y", alpha=0.3)
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```

```python
# Portfolio level improvement
plt.figure(figsize=(7,5))

plt.bar(
    ["Before", "After"],
    [before, after],
    color=[JPM_BROWN, JPM_TEAL],
    alpha=0.8
)

plt.title("Overall MAPE Before and After Calibration", color=JPM_CHARCOAL)
plt.ylabel("MAPE", color=JPM_CHARCOAL)
plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Product level improvement
plt.figure(figsize=(7,5))

plt.bar(
    compare_products.index,
    compare_products["MAPE_before"],
    color=JPM_BROWN,
    alpha=0.6,
    label="Before"
)

plt.bar(
    compare_products.index,
    compare_products["MAPE_after"],
    color=JPM_TEAL,
    alpha=0.7,
    label="After"
)

plt.title("MAPE by Product Before and After Calibration", color=JPM_CHARCOAL)
plt.ylabel("MAPE", color=JPM_CHARCOAL)

plt.grid(axis="y", alpha=0.3)
plt.legend(frameon=False)

plt.tight_layout()
plt.show()
```
